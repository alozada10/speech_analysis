```python
import unidecode
import glob
import spacy
import pandas as pd
import re
nlp = spacy.load('es_core_news_sm')
```

```python
def clean(text):
    text = text.lower().strip().replace('-','')
    text = unidecode.unidecode(text)
    return text
```

```python
def count_freq(text,
               top):

    frequency = {}
    pattern = re.findall(r'\b[a-z]{2,15}\b', text)
    for word in pattern:
        count = frequency.get(word,0)
        frequency[word] = count + 1
    temp= pd.DataFrame(frequency, range(1)).T.sort_values(0,ascending=False).head(20).reset_index().rename(columns={'index':'word',0:'freq'})
    return temp
```

```python
# freq
articles_path = 'data'

# Keeps titles and content
dates = []
contents = []

stop_w = nlp.Defaults.stop_words
new = ['de', 'la', 'el', 'en', 'no', 'los', 'que', 'las', 'por', 'es', 'hay', 'se', 'con','del', 'para', 'tiene']
for i in new: 
    stop_w.add(i)

data = None
for filename in glob.iglob(f'{articles_path}/**',recursive=True):
    if '.txt' in filename :
        date = int(filename.split('-')[4][:4])
        dates.append(date)
        
        with open(filename,'r', encoding='utf8') as file_n:
            text = nlp(clean(file_n.read()))
            text = [str(word) for word in text if (not str(word) in list(stop_w))&(not word.is_punct)&(not word.is_space)&(not word.is_digit)]
            text = ' '.join(text)
            temp = count_freq(text, 20)
            temp["year"] = [date]*temp.shape[0]
            data = pd.concat([data,temp])
```
